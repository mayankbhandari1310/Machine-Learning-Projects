```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating a black image, background for collage image
background_image = np.random.randint(0, 1, (430, 430, 3))
background_image.shape
plt.imshow(background_image)
plt.show()
```

```python
#read image
top_left = cv2.imread("ImageCollage dataset/top_left.jpg")
plt.imshow(top_left)
plt.title('Unaltered')
plt.show()

#color convert
rgb_top_left = cv2.cvtColor(top_left, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_top_left)
plt.title('color converted to rgb scheme')
plt.show()
rgb_top_left.shape

#resize
final_top_left = cv2.resize(rgb_top_left, (200, 200))
plt.imshow(final_top_left)
plt.title("resized and color converted")
plt.show()
final_top_left.shape
```

```python
#function to perform the previous alterations
def convert_image(image, size=(200, 200)):
    size = size
    image = cv2.resize(image, size)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image
```

```python
#reading images for collage
top_left = cv2.imread("ImageCollage dataset/top_left.jpg")
bottom_left = cv2.imread("ImageCollage dataset/bottom_left.jpg")
top_right = cv2.imread("ImageCollage dataset/top_right.jpg")
bottom_right = cv2.imread("ImageCollage dataset/bottom_right.jpg")
center = cv2.imread("ImageCollage dataset/center.jpeg") 
```

```python
#displaying all images
plt.imshow(top_left)
plt.show()
plt.imshow(bottom_left)
plt.show()
plt.imshow(top_right)
plt.show()
plt.imshow(bottom_right)
plt.show()
plt.imshow(center)
plt.show()
```

```python
#convert all the images
top_left = convert_image(top_left)
bottom_left = convert_image(bottom_left)
top_right = convert_image(top_right)
bottom_right = convert_image(bottom_right)
#center image is 100 X 100
center = convert_image(center, size=(100, 100))
```

```python
#display converted images
plt.imshow(top_left)
plt.show()
plt.imshow(bottom_left)
plt.show()
plt.imshow(top_right)
plt.show()
plt.imshow(bottom_right)
plt.show()
plt.imshow(center)
plt.show()
```

```python
collage_image = background_image
plt.imshow(collage_image)
```

```python
#insert images in collage
collage_image[10:210, 10:210, :] = top_left
plt.imshow(collage_image)
plt.show()
collage_image[220:420, 10:210, :] = bottom_left
plt.imshow(collage_image)
plt.show()
collage_image[10:210, 220:420, :] = top_right
plt.imshow(collage_image)
plt.show()
collage_image[220:420, 220:420, :] = bottom_right
plt.imshow(collage_image)
plt.show()
```

```python
center_start = int(430/2 - 60)
```

```python
#giving black outline to center image
center_image_with_black_border = np.random.randint(0, 1, (120, 120, 3))
center_image_with_black_border[10:110, 10:110, :] = center
plt.imshow(center_image_with_black_border)
plt.title('center image with black border')
```

```python
#final_collage image
collage_image[center_start:center_start+120, center_start:center_start+120, :] = center_image_with_black_border
#display final collage
plt.imshow(collage_image)
plt.title('collage')
plt.show()
```

```python
plt.imshow(collage_image)
plt.show()
```

```python
#resize to ((430*430), 3) shape
reshaped_image = np.reshape(collage_image, (430*430, 3))

#convert to dataframe and add column headers
import pandas as pd
final_image = pd.DataFrame(reshaped_image, columns=['r', 'g', 'b'])
print(final_image.head())
```

```python
#convert to csv files
final_image.to_csv('collage.csv', index=False)
```

```python
#opening and reading back the image
image = pd.read_csv('collage.csv')
image_array = image.values
print("image_array shape: ",image_array.shape)

#reshape back to  2D image
image = np.reshape(image_array, (430, 430, 3))
print("image shape: ",image.shape)
plt.imshow(image)
plt.title("collage")
plt.axis('off')
plt.show()
```
